```python
from operator import itemgetter 
```

```python
lines = tuple(open('../summary3/res.txt', 'r'))
```

```python
ress = []
for i in range(int(len(lines)/3)):
    a = lines[3*i].split("_")
    b = lines[3*i+1].split(",")
    a = list(itemgetter(*[3, 10, 11])(a))
    
    res = a + b
    res_ = []
    for el in res:
        el = el.replace('\n','')
        try:
            res_.append(float(el))
        except:
            res_.append(el)
    ress.append(res_)
    
```

```python
ress[1]
```

```python
import pandas as pd
df = pd.DataFrame(ress, columns=['data','lamda','split','crossval','acc','l_o_loss','a','b','obj','targetacc','gccacc','nogccacc'])
```

```python
idx = df.groupby(['data','split'])['acc'].transform(max) == df['acc']
```

```python
df_sub = df[idx].sort_values(['data','split'])
```

```python
prev = -20
final = []
for index, row in df_sub.iterrows():
    if prev == row['split']:
        pass
    else:
        final.append(row)
```

```python
pd.DataFrame(final).to_csv('lpn_random_train_and_val.csv')
```
