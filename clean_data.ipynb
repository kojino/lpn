```python
initial_baseline_accuracy = linqs_df.groupby(['data','split_seed']).agg({'parameter_seed':'min','initial_true_accuracy':'mean'}).reset_index()['initial_true_accuracy']
```

```python
idx = linqs_df.groupby(['data','split_seed'])['min_entropy_true_accuracy'].transform(max) == linqs_df['min_entropy_true_accuracy']
linqs_df[idx]

final_linqs = linqs_df.groupby(['data','split_seed','regularize','parameter_seed']).agg({'min_entropy_accuracy_diff':'max','min_entropy_true_accuracy':'max'}).reset_index()
```

```python
final_linqs['increase_from_baseline'] = final_linqs['min_entropy_true_accuracy'] - initial_baseline_accuracy.repeat(90).reset_index()['initial_true_accuracy']
```

```python
final_linqs.sort_values('increase_from_baseline',ascending=False)
```

```python
final_linqs['gcn_accuracy'] = gcn_df[gcn_df[0] == 'linqs_data/cora'][5]
final_linqs['gcn_val_accuracy'] = gcn_val_df[gcn_val_df[0] == 'linqs_data/cora'][3]
final_linqs['deepwalk_accuracy'] = deepwalk_df[2]
```

```python
final_linqs.sort_values('min_entropy_accuracy_diff',ascending=False)
```

```python
final_linqs['lp_closed_accuracy'] = nondeep_df[(nondeep_df['data'] == 'linqs/cora') & (nondeep_df['model'] == 'LP closed')].groupby(['data','split_seed']).agg({'accuracy':'max'}).reset_index()['accuracy']
```

```python
final_linqs['lp_iter10_weights1_accuracy'] = nondeep_df[(nondeep_df['data'] != 'linqs/pubmed') & (nondeep_df['model'] == 'LP Iter') & (nondeep_df['iter'] == '10') & (nondeep_df['weights'] == '1.0')].groupby(['data','split_seed']).agg({'accuracy':'max'}).reset_index()['accuracy']
```
