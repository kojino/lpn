```python
import numpy as np
import pandas as pd
```

```python
colnames = ['id','leave_k_out_loss','objective','unlabeled_loss','true_accuracy','entropy','micro_f1','macro_f1','regularize_val']
df = pd.read_csv('res_1/namespace.txt',delimiter='\t',header=None)

def rem_equal(el):
    _,_,val = el.partition('=')
    return val

df = df.iloc[:,[12,23,24,29]].applymap(rem_equal)
df.columns = ['loss_class_mass','parameter_seed','regularize','split_seed']
```

```python
file_names = ['initial', 'min_entropy', 'min_leave_k', 'max_accuracy', 'last']
for file in file_names:
    df_sub = pd.read_csv('res_1/'+file+'.txt',delimiter=' ',header=None)
    df_sub.columns = [file+'_'+colname for colname in colnames] 
    df = pd.concat([df,df_sub.iloc[:,1:]], axis=1)
```

```python
df['max_accuracy_diff'] = df['max_accuracy_true_accuracy'] - df['initial_true_accuracy']
df['min_entropy_accuracy_diff'] = df['min_entropy_true_accuracy'] - df['initial_true_accuracy']
df['last_accuracy_diff'] = df['last_true_accuracy'] - df['initial_true_accuracy']
sub_df = df.loc[:,['parameter_seed','split_seed','loss_class_mass','regularize','max_accuracy_true_accuracy','max_accuracy_diff','min_entropy_accuracy_diff','last_accuracy_diff','last_entropy','last_leave_k_out_loss','last_objective']]
```

```python
# sub_df.sort_values(['split_seed','loss_class_mass'], ascending=False).drop_duplicates(['split_seed','loss_class_mass'])
```

```python
# choosing the maximum accuracy for each split (upper limit)
sub_df.sort_values(['max_accuracy_diff'], ascending=False).drop_duplicates(['split_seed','parameter_seed','loss_class_mass'])[['split_seed','parameter_seed','loss_class_mass','regularize','max_accuracy_diff','max_accuracy_true_accuracy']]
```

```python
# choosing based on minimum entropy
sub_df[sub_df['loss_class_mass']=='1'].sort_values(['min_entropy_accuracy_diff'], ascending=False).drop_duplicates(['split_seed'])[['split_seed','loss_class_mass','regularize','min_entropy_accuracy_diff','max_accuracy_true_accuracy']]
```
