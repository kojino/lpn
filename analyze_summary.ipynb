```python
import os
fs = os.listdir("summary_bif")
```

```python
fss = []
for f in fs:
    if f[-5]=='5':
        fss.append(f)
```

```python
res
```

```python
ress = []
for f in fss:
    fp = open(f'summary_bif/{f}')
    acc = 0
    for i, line in enumerate(fp):
        if i in [200,400,600,800,1000]:
            acc += float(line.split(',')[-1].strip('\n'))
    acc = acc / 5.0
    fp.close()
    res = f.split('_')[2:-1]
    res = [float(el) for el in res]
    res.append(acc)
    if res[0] == 0.01:
        ress.append(res)
```

```python
# cols = ['lamda','split','t','acc','valacc','loss','a','b','obj']
cols = ['data','bif','keep_prob','split','acc','loss','a','b','obj']
# cols = ['bif','lamda','split','t','valacc']
```

```python
ress = []
for i in range(int(len(lines)/3)):
    a = lines[3*i].split('_')[3:]
    a[-1] = a[-1].split('.')[0]
    b = lines[3*i+1].split(',')
    res = a+b
    res = [el for el in res]
    ress.append(res)
```

```python
import pandas as pd
df = pd.DataFrame(ress, columns=cols)

# dfbif = dfbif.sort_values(by=['split','t'])
```
